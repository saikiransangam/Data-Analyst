```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
result = pd.read_csv('result.csv')
```

```python
result.info() 
```

```python
result.describe()
```

```python
globalavg_temp = result['globalavg_temp']
localavg_temp = result['localavg_temp']
```

```python
plt.plot(result['year'],globalavg_temp, label = 'Global')
plt.plot(result['year'], localavg_temp, label = 'Virginia Beach')
plt.legend()
plt.xlabel("years")
plt.ylabel("Celsius")
plt.title("Virginia Beach vs Global Temperature ")
```

```python
global_mavgtemp = result['globalavg_temp'].rolling(20).mean()
local_mavgtemp = result['localavg_temp'].rolling(20).mean()
```

```python
plt.plot(result['year'], global_mavgtemp, label = 'Global')
plt.plot(result['year'], local_mavgtemp, label = 'Virginia Beach')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Celsius")
plt.title("Virgina Beach vs Global moving average Temperature")
```

```python
plt.plot(result['year'], global_mavgtemp(result['globalavg_temp'], 10), label = '10 Moving Average Global Temp', alpha = 0.5)
plt.plot(result['year'], global_mavgtemp(result['globalavg_temp'], 30), label = '30 Moving Average Global Temp', alpha = 0.7)
plt.plot(result['year'], global_mavgtemp(result['globalavg_temp'], 50), label = '50 Moving Average Global Temp', alpha = 0.9)
plt.legend()
plt.xlabel("years")
plt.ylabel("Celsius")
plt.title(" Global Moving Average Temperature ")
plt.show()
```
